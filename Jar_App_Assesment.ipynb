```python
from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# upload the csv files
uploaded = files.upload()

# Loading the uploaded files into pd's
list_of_orders = pd.read_csv("List_of_Orders.csv")
order_details = pd.read_csv("Order_Details.csv")
sales_targets = pd.read_csv("Sales_Targets.csv")

# Display the top few rows of each dataset
print("List of Orders:")
print(list_of_orders.head())
print("\nOrder Details:")
print(order_details.head())
print("\nSales Targets:")
print(sales_targets.head())
```

**Sales Analysis:**

**Part 1: Sales and Profit Analysis**

Merge the List of Orders and Order Details datasets on the basis of Order ID.
Calculate the total sales (Amount) for each category across all orders.

```python
# Merging the two dataframes
merged_df = pd.merge(list_of_orders, order_details, on='Order ID', how='inner')

# merged dataframe
print("\nMerged Dataframe:")
print(merged_df.head())
```

```python
merged_df
```

```python
# Calculate total sales for each category
category_sales = merged_df.groupby('Category')['Amount'].sum()

# Display the results
print("\nTotal Sales by Category:")
category_sales
```

For each category, calculate the average profit per order and total profit margin
(profit as a percentage of Amount).

```python
# profit for each order
merged_df['Profit'] = merged_df['Amount'] - merged_df['Profit']

# profit margin
merged_df['Profit Margin'] = (merged_df['Profit'] / merged_df['Amount']) * 100

# Group by category and calculate average profit per order and total profit margin
category_profit_analysis = merged_df.groupby('Category').agg(
    {'Profit': 'mean', 'Profit Margin': 'sum'}
)

# Rename columns for better readability
category_profit_analysis = category_profit_analysis.rename(
    columns={
        'Profit': 'Average Profit per Order',
        'Profit Margin': 'Total Profit Margin'
    }
)

# Display the results
print("\nCategory-wise Profit Analysis:")
category_profit_analysis
```

Identify the top-performing and underperforming categories based on these
metrics. Also, suggest reasons for their performance diEerences.

```python
# Identify top and underperforming categories based on sales and profit

# Sort categories by total sales in descending order
top_performing_sales = category_sales.sort_values(ascending=False)
print("\nTop Performing Categories (Sales):")
print(top_performing_sales)

# Sort categories by average profit per order in descending order
top_performing_profit = category_profit_analysis.sort_values(
    by='Average Profit per Order', ascending=False
)
print("\nTop Performing Categories (Average Profit):")
print(top_performing_profit)


# Identify underperforming categories
underperforming_sales = category_sales.sort_values(ascending=True)
print("\nUnderperforming Categories (Sales):")
print(underperforming_sales)

underperforming_profit = category_profit_analysis.sort_values(
    by='Average Profit per Order', ascending=True
)
print("\nUnderperforming Categories (Average Profit):")
underperforming_profit
```

**Part 2: Target Achievement Analysis**

Using the Sales Target dataset, calculate the percentage change in target sales
for the Furniture category month-over-month.

```python
# Filter data for the Furniture category
furniture_targets = sales_targets[sales_targets['Category'] == 'Furniture']

# Create a new 'Month' column by extracting the month from 'Month of Order Date'
furniture_targets['Month'] = furniture_targets['Month of Order Date'].str.slice(0, 3)  # Extract first 3 characters for month

# Sort by month to ensure correct calculation of month-over-month change
furniture_targets = furniture_targets.sort_values('Month')

# Calculate the percentage change in target sales
furniture_targets['Percentage Change'] = furniture_targets['Target'].pct_change() * 100

# Display the results
print("\nMonth-over-Month Percentage Change in Target Sales for Furniture:")
print(furniture_targets[['Month', 'Target', 'Percentage Change']])
```

Analyse the trends to identify months with significant target fluctuations.
Suggest strategies for aligning target expectations with actual performance
trends.

```python
# Identify months with significant target fluctuations
significant_fluctuations = furniture_targets[furniture_targets['Percentage Change'].abs() > 10]  # Define a threshold (e.g., 10%)

print("\nMonths with Significant Target Fluctuations:")
print(significant_fluctuations[['Month', 'Target', 'Percentage Change']])
```

**Part 3: Regional Performance Insights**

From the List of Orders dataset, identify the top 5 states with the highest order
count. For each of these states, calculate the total sales and average profit.

```python
# Group by state and count orders
state_order_counts = merged_df.groupby('State')['Order ID'].count()

# Get the top 5 states with the highest order counts
top_5_states = state_order_counts.nlargest(5)

# Print the top 5 states
print("\nTop 5 States with Highest Order Counts:")
print(top_5_states)

# Calculate total sales and average profit for the top 5 states
top_5_state_analysis = merged_df[merged_df['State'].isin(top_5_states.index)].groupby('State').agg(
    {'Amount': 'sum', 'Profit': 'mean'}
)

# Rename columns for better readability
top_5_state_analysis = top_5_state_analysis.rename(
    columns={'Amount': 'Total Sales', 'Profit': 'Average Profit'}
)

# Display the results
print("\nTotal Sales and Average Profit for Top 5 States:")
top_5_state_analysis
```

```python
# Identify regional disparities and suggest areas for improvement.
print("\nRegional Performance Analysis (Top 5 States):")
print(top_5_state_analysis)

# Find the states with the lowest average profit
lowest_profit_states = top_5_state_analysis.sort_values(by='Average Profit').head(2)  # Showing the bottom 2 for illustration. Adjust as needed.
print("\nStates with Lowest Average Profit:")
print(lowest_profit_states)

# Prioritize regions for improvement: Focus on states with the lowest average profit.
print("\nPrioritization for Improvement:")
for state in lowest_profit_states.index:
    print(f"- Investigate reasons for low profitability in {state}. Consider targeted marketing campaigns, "
          f"price adjustments, or operational efficiency improvements.")
```
